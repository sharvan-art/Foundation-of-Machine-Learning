Pandas Exercise 1

```python
# Refrence Table: This is the raw data provided by the admin
# Name [Aman Saurav Karan Shubham Abhishek Missing Sameer]
# Age [22 23 Missing 22 25 24 26]
# Marks [85 Missing 88 90 56 72 Missing]
# Subject [Math Physics Math Chemistry Math Physics Chemistry]
# Note: "Missing" indicates data was not captured(NaN/None).

import pandas as pd
import numpy as np

# Create DataFrame
data = {
    "Name": ["Aman", "Saurav", "Karan", "Shubham", "Abhishek", None, "Sameer"],
    "Age": [22, 23, np.nan, 22, 25, 24, 26],
    "Marks": [85, np.nan, 88, 90, 56, 72, np.nan],
    "Subject": ["Math", "Physics", "Math", "Chemistry", "Math", "Physics", "Chemistry"]
}
```

```python
# Task 1.1: Create the DataFrame
# Construct a Pandas DataFrame named df that replicates the "Raw Data" table shown in the previous cell. Use None or np.nan for missing values

df = pd.DataFrame(data)
print(df)

# Task 1.2: Basic Inspection
# Once created: 1. Check the shape of the dataset. 
# 2. Display the data types of the columns. 
# 3. Generate a statistical summary(mean/min/max) for the numericals column.

print("Shape of the dataset:", df.shape)  #this will print the shape of the DataFrame in the form of (number of rows, number of columns)
print("\nData types of the columns:\n", df.dtypes)  #this will print the data types of the columns in the DataFrame
print("\nStatistical summary of the numerical columns:\n", df.describe())  #this will print the statistical summary of the numerical columns
```

```python
# Exercise 2: Selection & Aggregation
# Task: Filter and group the data. 
# 1.Selection: Create a view showing only the 'Name' and 'Marks' columns. 
df_selection = df[['Name', 'Marks']]
print(df_selection)

# 2.Filtering: Select students who scored above 80 marks. 
df_filtered = df[df['Marks'] > 80]
print(df_filtered)

# 3.Complex Filtering: Select students who are above 22 years of age and scored above 80 marks. 
df_complex_filtered = df[(df['Age'] > 22) & (df['Marks'] > 80)]
print(df_complex_filtered)

# 4.Aggregation: Count how many students are enrolled in each subject.
subject_count = df.groupby('Subject').size()
print(subject_count)
```

```python
# Exercise 3: Sorting & Ranking
# Task: Order the data to find rankers. 
# 1.Sort the DataFrame by 'Marks' in descending order. 
df_sorted = df.sort_values(by='Marks', ascending=False)
print(df_sorted)

# 2.Use a specific Pandas function to find the top 2 highest scorers (Scholorhip Candidates) without sorting the whole dataframe manually.
top_scorers = df.nlargest(2, 'Marks')
print(top_scorers)

# 3.Identify the student with the lowest marks (Remedial Candidate).
remedial_candidate = df.nsmallest(1, 'Marks')
print(remedial_candidate)

```

```python
# Exercise 4: Data Cleaning
# Task: Handle Missing Values
# 1. Calculate the total no. of missing values (Nan) in each column.
missing_values_count = df.isnull().sum()
print(missing_values_count)

# 2. Filter and display the rows where 'Marks' are missing.
missing_marks = df[df['Marks'].isnull()]
print(missing_marks)

# 3. Create a new dataframe called cleaned_df by removing any row that contains even a single missing value.
cleaned_df = df.dropna()
print(cleaned_df)

# 4. Compare the shape of cleaned_df vs original df.
print("Shape of original df:", df.shape)
print("Shape of cleaned_df:", cleaned_df.shape)
```

GroupBy in Pandas

```python
import pandas as pd
data={
    "Name": ["Aman", "Saurav", "Karan", "Shubham", "Abhishek", None, "Sameer"],
    "Age": [22, 23, None, 22, 25, 24, 26],
    "Marks": [85, None, 88, 90, 56, 72, None],
    "Subject": ["Math", "Physics", "Math", "Chemistry", "Math", "Physics", "Chemistry"]
}
df=pd.DataFrame(data)
print(df)
```

```python
# basic groupby
df.groupby('Subject')['Marks'].mean()  #this will group the DataFrame by the "Subject" column and calculate the mean of the "Marks" column for each group
```

```python
df.groupby('Subject')['Name'].count() #this will group the DataFrame by the "Subject" column and count the number of non-null values in the "Name" column for each group
```

```python
# Multiple Aggregations
df.groupby('Subject')['Marks'].agg(['mean', 'max', 'min'])  #this will group the DataFrame by the "Subject" column and calculate the mean, min and max of the "Marks" column for each group
```

```python
df.groupby('Subject')['Marks'].agg(['mean', 'max', 'min', 'count'])  #this will group the DataFrame by the "Subject" column and calculate the mean, min, max and count of the "Marks" column for each group
```

```python
# Group by Multiple Columns
df.groupby(['Subject', 'Age'])['Marks'].mean()  #this will group the DataFrame by both "Subject" and "Age" columns and calculate the mean of the "Marks" column for each group
```
