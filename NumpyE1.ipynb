```python
# Task 1.1 Create 1D array called cost price3 items [1500, 3000, 5000]
# Task 1.2 Create 2D array called daily sales [[10, 15, 8, 12], [5, 7, 4, 6], [2, 3, 2, 4]]
# Task 1.2 Print the shape and dimension of both arrays
import numpy as np
cost = np.array([1500, 3000, 5000])
daily_sales = np.array([[10, 15, 8, 12], [5, 7, 4, 6], [2, 3, 2, 4]])
print("Cost price array:", cost)
print("Daily sales array:\n", daily_sales)
print("Shape of cost price array:", cost.shape)
print("Dimension of cost price array:", cost.ndim)
print("Shape of daily sales array:", daily_sales.shape)
print("Dimension of daily sales array:", daily_sales.ndim)
```

```python
#  Task 2.1 (Grace Units)
# The manager decides to add **2 "bonus" units** to the daily sales of each item. Update the `daily_sales` array accordingly and print the updated array.
daily_sales += 2
print("Updated daily sales with bonus units:\n", daily_sales)
```

```python
#  Task 2.2 (Revenue)
# Calculate the total revenue for each day, 
# 1- Multiplying the cost_price with the total daily_sales matrix. 
# 2- Observe how numpy handles the multiplication of 1D and 2D arrays. give code
total_revenue = np.dot(cost, daily_sales)
print("Total revenue for each day:", total_revenue)
```

```python
# Task 2.3 (Scaling)
# Use np.sqrt() on the sales data to find the square root of all sales values for a statistical curve.
sqrt_sales = np.sqrt(daily_sales)
print("Square root of daily sales:\n", sqrt_sales)
```
