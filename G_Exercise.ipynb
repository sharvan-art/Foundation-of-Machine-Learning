Complete Code

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = {
"Monthly_Inc":[15000,18000,20000,22000,17000,19000,21000,23000,16000,20000,18000,22000,21000,24000,17000,19000,20000,23000,18000,21000],

"Online_Ord":[3,4,5,6,4,5,6,7,3,5,4,6,5,7,4,5,6,7,4,5],

"Spotify_Hrs":[1,2,3,4,2,3,4,5,1,3,2,4,3,5,2,3,4,5,3,4],

"Shop_Hours":[2,3,4,5,3,4,5,6,2,4,3,5,4,6,3,4,5,6,3,4],

"Money_Spent":[4000,4500,5000,6000,4200,4800,5500,6500,4100,5000,4600,6000,5300,6800,4300,4900,5200,6700,4500,5400]
}

df = pd.DataFrame(data)

print(df)
```

Statistical Output

```python
# Mean
print("Mean:\n",df.mean())

# Median
print("\nMedian:\n",df.median())

# Standard Deviation
print("\nStandard Deviation:\n",df.std())

# Correlation
print("\nCorrelation Matrix:\n",df.corr())
```

Visualisations

Scatter Plot (Income vs Spending)

```python
plt.scatter(df["Monthly_Inc"], df["Money_Spent"])

plt.title("Income vs Spending")
plt.xlabel("Monthly Income")
plt.ylabel("Money Spent")

plt.show()
```

Inference:

The scatter plot shows a positive relationship between monthly income and money spent. Customers with higher income tend to spend more on shopping. This indicates that income level influences customer spending behavior.

Histogram (Orders)

```python
plt.hist(df["Online_Ord"], bins=5)

plt.title("Distribution of Online Orders")
plt.xlabel("Orders")
plt.ylabel("Frequency")

plt.show()
```

Inference:

The histogram shows the distribution of online orders placed by customers. Most customers place between 4 and 6 orders. This indicates that the majority of customers have moderate online shopping activity.



Line Plot (Spotify Usage)

```python
plt.plot(df["Spotify_Hrs"], marker='o')

plt.title("Spotify Usage Trend")
plt.xlabel("Customer ID")
plt.ylabel("Spotify Hours")

plt.show()
```

Inference:

The line plot shows the trend of Spotify usage among customers. The listening hours vary between 1 and 5 hours. This indicates that users have different music listening habits but most of them spend a moderate amount of time on Spotify.
