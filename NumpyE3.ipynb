```python
# A university moniters electricity usage (in units) of 4 hostels over 7 days (Monday to Sunday).
# The data is stored in a 2D list as shown below:
```

```python
# Task A1: 
# Create a NumPy array electricity using np.random.randint:
# Values between 120 and 650
# Shape (4, 7)
import numpy as np
electricity = np.random.randint(120, 651, size=(4, 7)) 
print("Electricity usage (units) of 4 hostels over 7 days:\n", electricity)
```

```python
# Task A2:
# Print the following attributes of the array:
# Shape, Number of dimensions, Total number of elements, and Data type.
print("Shape of the array:", electricity.shape)
print("Number of dimensions:", electricity.ndim)
print("Total number of elements:", electricity.size)
print("Data type of the elements:", electricity.dtype)
```

```python
# Task B1: Hostel wise Metrics
# Find the average electricity usage per hostel, the maximum usage per hostel, and identify hostel has the highest average usage.
average_usage_per_hostel = np.mean(electricity, axis=1)
max_usage_per_hostel = np.max(electricity, axis=1)
hostel_with_highest_average = np.argmax(average_usage_per_hostel)
print("\nAverage electricity usage per hostel:", average_usage_per_hostel)
print("Maximum electricity usage per hostel:", max_usage_per_hostel)
print("Hostel with highest average usage:", hostel_with_highest_average)
```

```python
# Task B2: Day wise Metrics
# Calculate the total electricity consumption for each day and identify the day with the highest total campus consumption.
# Requirment: Students must choose the correct axis values for each operation.
total_consumption_per_day = np.sum(electricity, axis=0)
day_with_highest_consumption = np.argmax(total_consumption_per_day)
print("\nTotal electricity consumption per day:", total_consumption_per_day)
print("Day with the highest total campus consumption (0-indexed):", day_with_highest_consumption)
```

```python
# PART C - BROADCASTING & POLICY RULES
# Policy Rule: If electricity usage on any day is greater than 550 units, a 20 % surcharge is applied.
# Task C1
# Create a new array penalized_usage such that:
# Values > 550 -> multiplied by 1.20
# All other values remain unchanged.
# Constraints: Must use np.where. No loops allowed.
penalized_usage = np.where(electricity > 550, electricity * 1.20, electricity)
print("\nPenalized electricity usage (with surcharge applied where applicable):\n", penalized_usage)
```

```python
#  PART D - COMPARATIVE ANALYSIS
# Task D1: Thinking Task
# Compare each hostel's daily usage with its own weekly average. Create a boolean matrix where:
# True: Above hostel average
# False: Below or equal to hostel average
# Requirement: Requires broadcasting of a (4,) array across a(4, 7) array.
above_average_usage = electricity > average_usage_per_hostel[:, np.newaxis]
print("\nBoolean matrix indicating above average usage per hostel:\n", above_average_usage)
```

```python
# PART E - RESHAPING & DATA STRUCTURE
# Task E1: Reshape the electricity data into a single row representing weekly usage (Shape 1x28).
reshaped_electricity = electricity.reshape(1, -1)
print("\nReshaped electricity data (1x28):\n", reshaped_electricity)
```

```python
# Task E2
# Reshape the same data into 2 rows and 14 columns.Ensure no data is lost or duplicated.
reshaped_electricity_2 = electricity.reshape(2, 14)
print("\nReshaped electricity data (2x14):\n", reshaped_electricity_2)
```

```python
# PART F - MEMORY SAFETY
# Task F1: Copy vs View
# Create ref_backup=electricity
# Create safe_backup=electricity.copy()
# Modify one value in the original electricity array.
# Observe which backups change.
# Explaination Reuired: Why is ref_backup risky? Why is safe_backup safer?
ref_backup = electricity
safe_backup = electricity.copy()
# Modify one value in the original electricity array
electricity[0, 0] = 999
print("\nOriginal electricity array after modification:\n", electricity)
print("\nReference backup (ref_backup):\n", ref_backup)
print("\nSafe backup (safe_backup):\n", safe_backup)
```
