Data Visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = {
    "Name": ["Aman","Riya","Rahul","Sneha","Arjun","Priya","Karan","Neha"],
    "Age": [22,25,24,23,26,27,28,22],
    "Gender": ["M","F","M","F","M","F","M","F"],
    "Marks": [85,90,78,88,92,95,80,87]
}

df = pd.DataFrame(data)

print(df)
```

Line Plot for marks trend

```python
plt.plot(df["Marks"])

plt.title("Marks Trend")
plt.xlabel("Student")   # correct function
plt.ylabel("Marks")

plt.show()
```

Bar plot for comparision

```python
plt.bar(df["Name"], df["Marks"])
plt.title("Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Marks")

plt.show()
```

Histograme for age distribution

```python
plt.hist(df["Age"], bins=5)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

Scatter Plot

Relationship between Age and Marks

```python
plt.scatter(df["Age"], df["Marks"])

plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")

plt.show()
```

Pie Chart for Gender Distribution

```python
gender_counts=df["Gender"].value_counts()

plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

Customizing Plots

Example with colors and grid

```python
plt.bar(df["Name"], df["Marks"], color="green")

plt.title("Student Marks")
plt.xlabel("Student Name")
plt.ylabel("Marks")

plt.grid(True)

plt.show()
```

Subplot

Show two graphs together

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(df["Age"])
plt.title("Age Distribution")

plt.subplot(1,2,2)
plt.bar(df["Name"], df["Marks"])
plt.title("Marks of Students")

plt.show()
```
